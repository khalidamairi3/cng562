# binary classification for small fires

###### i will add a column to the data frame and i prefered to have 3 classes which are { "no fire":0, "small fire":1,"big fire":2} .. so it won't be a binary classification. 
###### for the threshold i chose 75 to be the point of differentiating the small fire from the big one

```python
data['small.fires']=pd.Series([])
inputs = data# the inputs are all columns in the dataset except income

inputs=inputs.values



for i in range (0,517):
    if inputs[i][12]>= 70:
        inputs[i][13]=2
    elif inputs[i][12]> 0:
        inputs[i][13]=1
    else:
        inputs[i][13]=0

```

```python
data=pd.DataFrame({'X':inputs[:,0].tolist(), 'Y':inputs[:,1].tolist(),'month':inputs[:,2].tolist(),'day':inputs[:,3].tolist(),'FFMC':inputs[:,4].tolist(),'DMC':inputs[:,5].tolist(),'DC':inputs[:,6].tolist(),'ISI':inputs[:,7].tolist(),'temp':inputs[:,8].tolist(),'RH':inputs[:,9].tolist(),'wind':inputs[:,10].tolist(),'rain':inputs[:,11].tolist(), 'area':inputs[:,12].tolist(), 'small.fires':inputs[:,13].tolist()})
data 
```

```python
inputs = data.drop('area',axis='columns')
inputs = inputs.drop('small.fires',axis='columns')

target = data['small.fires']
inputs=pd.get_dummies(inputs, sparse=True) 
inputs=inputs.values
target=target.values


x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.1,random_state=1)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
# predict for x_test 
y_pred=model.predict(x_test)
#calculate the accuracy of prediction by comparing the results
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(x_train, y_train)  
y_pred = classifier.predict(x_test) 
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
y_pred=logistic.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.model_selection import cross_val_score,KFold

logistic = LogisticRegression()
classifier = KNeighborsClassifier(n_neighbors=5)
model = tree.DecisionTreeClassifier()



scores_logistic = cross_val_score(logistic, inputs, target, cv=10) 

scores_tree = cross_val_score(model, inputs, target, cv=10)

scores_knn = cross_val_score(classifier, inputs, target, cv=10)


print("Accuracy of logistic regression: %0.2f " % (scores_logistic.mean()))
print("Accuracy of decision tree : %0.2f " % (scores_tree.mean()))
print("Accuracy of knn : %0.2f " % (scores_knn.mean()))
```
